```python
import re 
# 한글만 추출하는 함수 
def test(s):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣+]') # 한글과 띄어쓰기를 제외한 모든 글자
    # hangul = re.compile('[^ \u3131-\u3163\uac00-\ud7a3]+')  # 위와 동일
    result = hangul.sub('', s) # 한글과 띄어쓰기를 제외한 모든 부분을 제거
    return(result)
```

```python
# 딕셔너리 자료형으로 저장 
count = 0
labeling_dic = {}
labeling_key = 0
for str in result_list : 
    tokens = str.split('\n')
    room_list = []
    for token in tokens :
        if(len(test(token))> 0) :  
            room_list.append(test(token))
    labeling_dic[labeling_key] = room_list
    labeling_key  += 1
labeling_dic
```
